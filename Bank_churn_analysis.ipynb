```python
# Business Question #1
# Does Age Impact Churn?
# "Are older customers leaving more often than younger customers?"
df.groupby('Exited')['Age'].describe()
```

```python
# Visualization
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(x='Exited', y='Age', data=df)
plt.title('Age vs Churn')
plt.show()
```

```python
# Age group
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[18,30,40,50,60,100],
    labels=['18-30','31-40','41-50','51-60','60+']
)

pd.crosstab(
    df['Age_Group'],
    df['Exited'],
    normalize='index'
)*100
```

```python
# Business Question #2
# Does Geography Impact Churn?
# "Which country is responsible for most customer losses?".
pd.crosstab(
    df['Geography'],
    df['Exited'],
    normalize='index'
)*100
```

```python
# visualization
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='Geography',
    y='Exited',
    estimator=lambda x: x.mean()*100
)

plt.ylabel('Churn Rate (%)')
plt.title('Churn Rate by Geography')
plt.show()
```

```python
# Business Question #6
# Does Account Balance Affect Churn?
# Are customers with higher balances more likely to leave?

# Average Balance by Churn
df.groupby('Exited')['Balance'].describe()
```

```python
# Boxplot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(x='Exited', y='Balance', data=df)
plt.title('Balance vs Churn')
plt.show()
```

```python
# Create Balance Segments
df['Balance_Group'] = pd.qcut(
    df['Balance'].rank(method='first'),
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)
```

```python
# Churn by Balance Group
pd.crosstab(
    df['Balance_Group'],
    df['Exited'],
    normalize='index'
)*100
```
